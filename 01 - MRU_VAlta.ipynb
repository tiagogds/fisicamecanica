# Análise de um experimento - MRU

Nesta aula iremos verificar experimentalmente alguns aspectos do movimento com velocidade constante.

Neste tipo de movimento, a posição de um objeto é bem descrita através da equação:

\begin{equation}
x(t)=vt+x_0
\end{equation}

onde $x$ é a posição atual da partícula, $t$ é o tempo, $v$ é a velocidade inicial (constante) e $x_0$ a posição no tempo inicial, *i.e.*, em $t=0s$.
De um ponto de vista puramente matemático, podemos enxergar a equação acima como uma equação de primeiro grau onde a variável independente é o tempo $t$, a inclinação (ou coeficiente angular) é a velocidade e o coeficiente linear é a posição inicial.

O vídeo mostrado foi analisado utilizando-se um o *software* tracker. Neste *software* foram extraídas as seguintes informações:

```python
if True:
    t = []
    x = []
    vx = []
    arq = open('01 - MRU_VAlta.dat')
    arq = arq.read()
    arq = arq.replace(',','.')
    arq = arq.split('\n')
    
    for i in range(1,len(arq)-1):
        l = arq[i]
        l = l.split(' ')
        if l[0]:
            t.append(float(l[0]))
            x.append(float(l[1]))
            vx.append(float(l[2]))

    t = array(t)
    x = array(x)
    vx = array(vx)

    print("t (s)\tx (m)\tv(m/s)")
    for i in range(0,10):
        print("%.2f\t%.2f\t%.2f" %(t[i],x[i],vx[i]))
```

```python
if True:
    l1 = plt.plot(t,x,label='Medido')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição (m)')
    coef = polyfit(t,x,1)
    fit = coef[0] * t + coef[1]
    l2 = plt.plot(t,fit,label='Teórico')
    plt.legend()
    plt.show()
    coef[0] = coef[0] * 100
    coef[0] = round(coef[0])
    coef[0] = coef[0]/100.0
    print('Equação do movimento: ')
    print('x = %.2ft + %.2f'%(coef[0],coef[1]))
```

```python
if True:
    l1 = plt.plot(t,vx,label='Medido')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição (m)')
    coef = polyfit(t,vx,1)
    fit = coef[0] * t + coef[1]
    l2 = plt.plot(t,fit, label='Teórico')
    plt.legend()
    plt.show()
    print('Equação da reta encontrada: ')
    print('v = %.2ft + %.2f'%(abs(coef[0]),coef[1]))
```

Perceba que o coeficiente angular está zerado e que o coeficiente linear é muito próximo do valor encontrado na equação para x.
